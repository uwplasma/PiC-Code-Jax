```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# File path
file_path = 'D:\\Desktop\\code\\PiC-Code-Jax-main\\new\\1000_steps_test.csv'

# Read the data from the CSV file
df = pd.read_csv(file_path)

# Check for missing values
if df.isnull().values.any():
    print("Warning: There are missing values in the data. Filling missing values with zero.")
    df = df.fillna(0)

# Create a pivot table to ensure data is properly shaped
pivot_df = df.pivot(index='k', columns='A', values='func_values')

# Check if pivot table contains any NaN values and fill them if necessary
if pivot_df.isnull().values.any():
    print("Warning: Pivot table contains NaN values. Filling NaN values with zero.")
    pivot_df = pivot_df.fillna(0)


# Extract the grid data from the pivot table
A_grid = pivot_df.columns.values*10**2
k_grid = pivot_df.index.values
#func_values = pivot_df.values

#A_grid = np.log(pivot_df.columns.values)
func_values = np.log(pivot_df.values)
print(np.min(pivot_df.values))

# Create a meshgrid for A and k
A_mesh, k_mesh = np.meshgrid(A_grid, k_grid)

# Create a 2D density plot
fig, ax = plt.subplots()
contour = ax.contourf(A_mesh, k_mesh, func_values, cmap='viridis')

# Add labels and title
ax.set_xlabel('A($10^{-2}$)')
ax.set_ylabel('k')
ax.set_title('Density Plot of electric field energy in log scale')

# Add a color bar which maps values to colors
fig.colorbar(contour, shrink=0.5, aspect=5)

plt.show()
```
